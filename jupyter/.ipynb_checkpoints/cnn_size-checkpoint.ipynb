```python
import pandas as pd
import seaborn as sns
import csv

df = pd.read_csv("../data/cnn_size_std_mode.csv")
```

```python
df
```

```python
palette = sns.cubehelix_palette(start=.5, rot=-.5, as_cmap=True)

ax = sns.displot(data=df, 
                x="total_power_avg", 
                hue = 'usb_type', 
                # kind="kde",
                fill=True, 
                height=2, 
                 aspect=3, palette=[palette.colors[60],palette.colors[-1]])

ax.set(xlabel='Overall Power (W)')
ax._legend.set_title("USB Type")
new_labels = ['V2', 'V3']
for t, l in zip(ax._legend.texts, new_labels):
    t.set_text(l)
ax.figure.savefig("../output/power_histogram.png")
```

```python
g = sns.catplot(
    data=df, kind="bar",
    x="block_type", y="inference_time_per_input", hue="usb_type", fill=True, alpha=.6, 
                height=2, 
                 aspect=3, palette=[palette.colors[60],palette.colors[-1]], linewidth=1, ec='k', ci=None
)
g.ax.set_ylabel("Inference\nDuration(s)")
g.ax.set_xlabel("Block Type")
g._legend.set_title("USB Type")
new_labels = ['V2', 'V3']
for t, l in zip(g._legend.texts, new_labels):
    t.set_text(l)

g.figure.savefig("../output/inference_time_usb.png")
```

```python
g = sns.catplot(
    data=df, kind="bar",
    x="block_type", y="total_power_avg", hue="usb_type", fill=True, alpha=.6, 
                height=2, 
                 aspect=3, estimator='mean', palette=[palette.colors[60],palette.colors[-1]], linewidth=1, ec='k', ci=None
)
g.ax.set_ylabel("Power (W)")
g.ax.set_xlabel("Block Type")
g._legend.set_title("USB Type")
new_labels = ['V2', 'V3']
for t, l in zip(g._legend.texts, new_labels):
    t.set_text(l)

g.figure.savefig("../output/power_usb.png")
```

```python
g = sns.catplot(
    data=df, kind="bar",
    x="block_type", y="total_power_avg", hue="usb_type", alpha=.6, 
                height=3.5, 
                 aspect=2, estimator='median', palette=[palette.colors[60],palette.colors[-1]], ec='k'
)
```

```python
import matplotlib.pyplot as plt
fig, axes = plt.subplots(3, 2, figsize=(7, 9))
dtypes = ["fullconv", "separable", "glu"]
for i in [2,3]:
    for j in range(len(dtypes)):
        ax = sns.scatterplot(ax=axes[j,i-2],
            data=df[(df["usb_type"]==f"usb{i}") & (df["block_type"]==dtypes[j])], 
            x="total_filters", y="cpu_utilization_avg", hue="number_of_layers", legend="full", palette=palette)
        ax.set_title(f"{dtypes[j]} CNN on USB V{i}")
        ax.set_ylabel("CPU Utilization")
        ax.set_xlabel("Filters")
        if i > 2 and j < 1:
            ax.legend(ncols=2).set_title("# Layers")
        else:
            ax.legend().remove()
        ax.set(ylim=(0, 50))
fig.tight_layout()
fig.savefig("../output/cpu_util_cnn_size.png")
```

```python
fig, axes = plt.subplots(3, 2, figsize=(7, 9))
dtypes = ["fullconv", "separable", "glu"]
for i in [2,3]:
    for j in range(len(dtypes)):
        ax = sns.scatterplot(ax=axes[j,i-2],
            data=df[(df["usb_type"]==f"usb{i}") & (df["block_type"]==dtypes[j])], 
            x="total_filters", y="total_power_avg", hue="number_of_layers", legend="full", palette=palette)
        ax.set_title(f"{dtypes[j]} CNN on USB V{i}")
        ax.set_ylabel("Power (W)")
        ax.set_xlabel("Filters")
        if i > 2 and j < 1:
            ax.legend(ncols=2).set_title("# Layers")
        else:
            ax.legend().remove()
        ax.set(ylim=(4, 9))
fig.tight_layout()
fig.savefig("../output/power_cnn_size.png")
```

```python
plt.figure(figsize=(7,3.5))
ax = sns.scatterplot(
    data=df[(df["usb_type"]=="usb3") & (df["block_type"]=="fullconv")], 
    x="total_filters", y="inference_time_per_input", hue="number_of_layers", legend="full", palette=palette)
sns.move_legend(ax, "upper left", bbox_to_anchor=(1, 1))
```

```python
plt.figure(figsize=(3,7))
df["Layers"] = df["number_of_layers"]
df["Filters \nper layer"] = df["filters_per_layer"]
df["Joules per Input"] = df["joules_per_input"]
df["Filters"] = df["total_filters"]

ax = sns.scatterplot(
    data=df[(df["usb_type"]=="usb2") & (df["block_type"]=="fullconv") & (df["filters_per_layer"]>=256)], 
    x="Filters", y="Joules per Input", hue="Layers", palette=palette, size="Filters \nper layer",sizes=(10, 200))
ax.set_title(f"Filters>=256")
ax.figure.savefig("../output/joules_big.png")
```

```python
ax = sns.scatterplot(df[(df["usb_type"] == "usb2")& (df["block_type"]=="fullconv")],x="total_filters", y="tpu_power_avg",  color=palette.colors[120],
                    label="TPU")
ax = sns.scatterplot(df[(df["usb_type"] == "usb2")& (df["block_type"]=="fullconv")],x="total_filters", y="cpu_power_avg", ax=ax, color=palette.colors[-1], 
                     label="CPU")
ax.set(ylim=(-0.2, 3))
ax.set_ylabel("Power(W)")
ax.set_xlabel("Filters")

ax.set_title("CPU vs TPU power for USB V2")
ax.figure.set_size_inches(6,2.5)
ax.legend().set_title("Device")
ax.figure.tight_layout()
ax.figure.savefig("../output/power_cnn_size_usb2.png")
```
