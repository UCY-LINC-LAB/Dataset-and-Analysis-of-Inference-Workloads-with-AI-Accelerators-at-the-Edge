```python
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_percentage_error

data = pd.read_csv('../data/cpu_util_power.csv')

x = data['cpu_utilization'].values
y = data['actual_power_consumption'].values

z = np.polyfit(x, y, 1)
p = np.poly1d(z)

y_pred = p(x)

palette = sns.cubehelix_palette(start=.5, rot=-.5, as_cmap=True)
fig = plt.figure(figsize=(6.5, 2))
plt.scatter(x, y, alpha=0.5, color=palette.colors[60])
plt.plot(x, y_pred, color='red', label='Linear Regression Fitting')

plt.xlabel("CPU Utilization (%)")
plt.ylabel("Power (W)")
plt.title('CPU Utilization vs Power Consumption')

mse = mean_squared_error(y, y_pred)
mape = mean_absolute_percentage_error(y, y_pred)
print(f'Mean Squared Error: {mse:.4f}')
print(f'Mean Absolute Percentage Error: {mape:.4f}')
np.save('../models/cpu-regression-model.npy', p)
print("Coefficients:", z)

plt.legend()
fig.tight_layout()
plt.savefig("../output/cpu-power.png")
plt.show()
```
